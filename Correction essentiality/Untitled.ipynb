```python
df_sapiens = pd.read_csv('Mus musculus_consolidated.csv').drop(['locus', 'datasets', 'datasetIDs', 'essentiality status'], 1)
df_sapiens.head()
```

```python
genes = []
for k in df['Pair']:
    g1, g2 = k.split('/')
    if g1 not in genes:
        genes.append(g1)
    if g2 not in genes:
        genes.append(g2)
genes = sorted(genes)
```

```python
lookup_ess = {}
for line in array(df_sapiens):
    name, ess = line
    if type(name) is float: continue;
    lookup_ess[name.upper()] = ess
```

```python
result_s = {}
for g in genes:
    if g in lookup_ess:
        result_s[g] = lookup_ess[g]
    else:
        result_s[g] = 'N/A'
        print(g, 'not found.')
```

```python
for key in result_s:
    x = result_s[key]
    if x == 'Essential':
        result_s[key] = 1
    elif x == 'Nonessential':
        result_s[key] = 0
```

```python
new_essA, new_essB = [], []
for pair in df['Pair']:
    g1, g2 = pair.split('/')
    new_essA.append(result_s[g1])
    new_essB.append(result_s[g2])
new_essA = array(new_essA)
new_essB = array(new_essB)
```

```python
df2 = pd.read_csv('dida_v2_full.csv', index_col=0)
new_essA[new_essA == 'N/A'] = 0.67
new_essB[new_essB == 'N/A'] = 0.62
df2['EssA'] = new_essA
df2['EssB'] = new_essB
```

```python
df2.to_csv('dida_v2_full_newess.csv')
```
